```python
import torch
from datasets import load_dataset, Dataset
from transformers import (
    AutoTokenizer,
    AutoModelForCausalLM,
    Trainer,
    TrainingArguments,
)
from peft import get_peft_model, LoraConfig, TaskType
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
import numpy as np
```

```python
model_name = "deepseek-ai/deepseek-coder-1.3b-base"

tokenizer = AutoTokenizer.from_pretrained(model_name)
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype=torch.bfloat16,
    device_map="auto",        # ✅ uses both GPUs automatically
    low_cpu_mem_usage=True
)
```

```python

lora_config = LoraConfig(
    r=8,                     # ✅ fixed rank = 8
    lora_alpha=16,
    lora_dropout=0.05,
    target_modules=["q_proj", "k_proj", "v_proj", "o_proj"],  
    task_type=TaskType.CAUSAL_LM,
)
model = get_peft_model(model, lora_config)
model.print_trainable_parameters()
```

```python
dataset = load_dataset("wikitext", "wikitext-2-raw-v1")

train_texts = dataset["train"]["text"][:8000]
val_texts = dataset["validation"]["text"][:1000]
test_texts = dataset["test"]["text"][:1000]

def tokenize_data(texts):
    enc = tokenizer(
        texts,
        truncation=True,
        padding="max_length",
        max_length=128,
        return_tensors="pt"
    )
    enc["labels"] = enc["input_ids"].clone()
    return Dataset.from_dict({k: v.tolist() for k, v in enc.items()})
```

```python
train_dataset = tokenize_data(train_texts)
val_dataset = tokenize_data(val_texts)
test_dataset = tokenize_data(test_texts)
```

```python
def compute_metrics(eval_pred):
    logits, labels = eval_pred
    logits = torch.tensor(logits)
    labels = torch.tensor(labels)

    # Compute loss & perplexity
    shift_logits = logits[..., :-1, :].contiguous()
    shift_labels = labels[..., 1:].contiguous()
    loss_fct = torch.nn.CrossEntropyLoss(ignore_index=-100)
    loss = loss_fct(shift_logits.view(-1, shift_logits.size(-1)), shift_labels.view(-1))
    perplexity = torch.exp(loss)

    # Decode text
    preds = torch.argmax(logits, dim=-1)
    decoded_preds = tokenizer.batch_decode(preds, skip_special_tokens=True)
    decoded_labels = tokenizer.batch_decode(labels, skip_special_tokens=True)

    # Compute BLEU
    smoothie = SmoothingFunction().method4
    bleu_scores = [
        sentence_bleu([ref.split()], pred.split(), smoothing_function=smoothie)
        for pred, ref in zip(decoded_preds, decoded_labels)
    ]

    return {
        "loss": loss.item(),
        "perplexity": perplexity.item(),
        "bleu": float(np.mean(bleu_scores))
    }
```

```python
training_args = TrainingArguments(
    output_dir="./deepseek_wiki_lora_baseline",
    num_train_epochs=6,
    per_device_train_batch_size=2,
    per_device_eval_batch_size=2,
    gradient_accumulation_steps=2,
    learning_rate=2e-5,
    evaluation_strategy="epoch",
    logging_steps=10,
    bf16=True,
    save_strategy="no",
    report_to="none",
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    tokenizer=tokenizer,
    compute_metrics=compute_metrics,
)
```

```python
print("\n🚀 Starting Standard LoRA (r=8) fine-tuning on WikiText...")
trainer.train()
```

```python
print("\n🔍 Final Evaluation on Test Split (Standard LoRA)...")
metrics = trainer.evaluate(test_dataset)
print(metrics)
```
